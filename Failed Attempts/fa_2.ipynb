```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('data.csv')
```

```python
df
```

```python
df.info(verbose=True, show_counts=True)
```

```python
for col in df.columns:
    print(f"Column: {col}")
    print(df[col].unique())
    print("-" * 40)
```

```python
print(df.isnull().sum())
```

```python
df = df[(df['manufacture_year'] >= 1886) & (df['manufacture_year'] <= 2025)]
```

```python
df.info(verbose=True, show_counts=True)
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['manufacture_year'], bins=10, kde=True, color='skyblue')
plt.title('Distribution of Manufacture Year')
plt.xlabel('Year')
plt.ylabel('Number of Cars')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['mileage'], bins=50, kde=True, color='salmon')
plt.title('Distribution of Mileage')
plt.xlabel('Mileage (km)')
plt.ylabel('Number of Cars')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
top_makers = df['maker'].value_counts().nlargest(10)
sns.barplot(x=top_makers.index, y=top_makers.values, palette='coolwarm')
plt.title('Top 10 Car Makers')
plt.xlabel('Car Maker')
plt.ylabel('Number of Cars')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(12, 6))
top_models = df['model'].value_counts().nlargest(10)
sns.barplot(x=top_models.index, y=top_models.values, palette='viridis')
plt.title('Top 10 Car Models')
plt.xlabel('Car Model')
plt.ylabel('Number of Cars')
plt.xticks(rotation=45)
plt.show()
```
